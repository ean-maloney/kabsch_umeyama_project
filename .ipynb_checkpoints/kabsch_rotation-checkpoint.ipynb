```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import itertools
```

```python
color_weight = [0.2125, 0.7154, 0.0721]; # LUMA-REC.709
```

```python
img_P = mpimg.imread('page_straight.jpg')
img_Q = mpimg.imread('page_rotated_45.jpg')
img_L = mpimg.imread('page_rotated_270.jpg')
```

```python
def load_images(imgs):
    out = []
    
    for img in imgs:
        white_coords = polarize_img(img, color_weight)
        corners = np.array(get_corners(white_coords))
        
        out.append(corners)
        
    return out
```

```python
def polarize_img(img, color_weight=color_weight, show=False):
    img_gray = np.dot(img[..., 0:3], color_weight)
    img_bw = np.where(img_gray < 150, 0, 1)
    
    if show:
        plt.imshow(img_bw, cmap='gray')
        return 
    
    y_white, x_white = img_bw.nonzero()
    white_coords = np.column_stack((x_white, y_white))
    
    return white_coords
```

```python
def get_corners(white_coords):
    y_max = white_coords[:, 1].max()
    
    if np.count_nonzero(white_coords[:, 1] == y_max) > 1:
        corners = get_corners_parallel(white_coords)
        
    else: 
        corners = get_corners_skew(white_coords)
        
    return corners
```

```python
def get_corners_parallel(white_coords):
    y_max, y_min = white_coords[:, 1].max(), white_coords[:, 1].min()
    y_max_coords, y_min_coords = white_coords[white_coords[:, 1] == y_max], white_coords[white_coords[:, 1] == y_min]
    
    corner_tl = (y_min_coords[:, 0].min(), y_min)
    corner_tr = (y_min_coords[:, 0].max(), y_min)
    corner_bl = (y_max_coords[:, 0].min(), y_max)
    corner_br = (y_max_coords[:, 0].max(), y_max)
    
    return corner_tl, corner_tr, corner_bl, corner_br    
```

```python
def get_corners_skew(white_coords):
    x_max, x_min = white_coords[:, 0].max(), white_coords[:, 0].min()
    y_max, y_min = white_coords[:, 1].max(), white_coords[:, 1].min()
    
    corner_0 = (x_max, white_coords[white_coords[:, 0] == x_max][0][1])
    corner_1 = (x_min, white_coords[white_coords[:, 0] == x_min][0][1])
    corner_2 = (white_coords[white_coords[:, 1] == y_max][0][0], y_max)
    corner_3 = (white_coords[white_coords[:, 1] == y_min][0][0], y_min)
    
    return corner_0, corner_1, corner_2, corner_3
```

```python
def get_points(A):
    points_x = []
    points_y = []
    
    for row in A:
        points_x.append(row[0])
        points_y.append(row[1])
    
    return list(zip(points_x, points_y))
```

```python
def plot_points(A, color='k'):
    points = get_points(A)
    
    for item in points:
        plt.plot(item[0], item[1], f'{color}o', markersize=10)
```

```python
def plot_lines(A, color='k'):
    points = get_points(A)
    lines = [[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]
    
    style = '-'
    
    for item in lines:
        point_a = points[item[0]]
        point_b = points[item[1]]
        
        plt.plot([point_a[0], point_b[0]], [point_a[1], point_b[1]], f'{color}{style}')
        
#         if style == ':':
#             style = '-'
        
#         else:
#             style = ':'
```

```python
p_, q_, l_ = load_images([img_P, img_Q, img_L])
```

```python
p_
```

```python
ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()
plot_points(p_, 'b')
plot_lines(p_, 'b')

plot_points(q_, 'r')
plot_lines(q_, 'r')

plot_points(l_, 'y')
plot_lines(l_, 'y')
```

```python
def center_matrix(A):
    centroid = (A[:, 0].mean(), A[:, 1].mean())
    
    return A - centroid
```

```python
Q = center_matrix(q_)
P = center_matrix(p_)
L = center_matrix(l_)
```

```python
ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()
plot_points(P, 'b')
plot_lines(P, 'b')

plot_points(Q, 'r')
plot_lines(Q, 'r')

plot_points(L, 'y')
plot_lines(L, 'y')
```

```python
def get_kabsch_rotation(A, B):
    '''
    Rotation of A into B
    '''
    
    H = A.T @ B

    U, S, V = np.linalg.svd(H, full_matrices=True)
    
    d_ = np.linalg.det(V) * np.linalg.det(U)

    if d_ >= 0:
        I = np.eye(2)
    
    else: 
        I = np.array([[1, 0], [0, -1]])
       
    R = U @ I @ V.T
    
    return R
```

```python
R_1 = get_kabsch_rotation(P, Q)

ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()

print(np.linalg.norm(Q - P @ R_1))

plot_points(P)
plot_lines(P)

plot_points(Q, 'r')
plot_lines(Q, 'r')

plot_points(P @ R_1, 'b')
plot_lines(P @ R_1, 'b')
```

```python
R_2 = get_kabsch_rotation(Q, L)

ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()

print(np.linalg.norm(L - Q @ R_2))

plot_points(Q, 'k')
plot_lines(Q, 'k')

plot_points(L, 'y')
plot_lines(L, 'y')

plot_points(Q @ R_2, 'r')
plot_lines(Q @ R_2, 'r')
```

```python
R_3 = get_kabsch_rotation(L, P)

ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()

print(np.linalg.norm(P - L @ R_3))

plot_points(L, 'k')
plot_lines(L, 'k')

plot_points(P, 'b')
plot_lines(P, 'b')

plot_points(L @ R_3, 'y')
plot_lines(L @ R_3, 'y')
```

```python
# Obviously R_3 isn't miniminizing the objective function, because multiplying R by [[-1, 0], [0, -1]] is better.
R_0 = get_kabsch_rotation(L, P) @ np.array([[-1, 0], [0, -1]])

ax = plt.gca()
ax.set_aspect('equal')
ax.invert_yaxis()
                                        
print(np.linalg.norm(P - L @ R_0))

plot_points(L, 'k')
plot_lines(L, 'k')

plot_points(P, 'b')
plot_lines(P, 'b')

plot_points(L @ R_0, 'y')
plot_lines(L @ R_0, 'y')
```
